```python
def a_star(start,stop):
  visited = 0
  city = []
  city.append(start)
  while city[-1] != stop :
    tetangga = data[data['Tetangga'] == city[-1]]
    tetangga['fn'] = tetangga['heuristik']+tetangga['jarak']+visited
    city_terdekat=tetangga.loc[tetangga['fn'].idxmin(),'Nama Kota']
    
    city.append(city_terdekat)
    print(city,'\n')
  return city
```

```python
a = a_star('Arad','Bucharest')
```
